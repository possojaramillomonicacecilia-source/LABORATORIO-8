```python
import random

def generar_numero():
    """Generates a random integer between 1 and 100."""
    return random.randint(1, 100)

print("¡Bienvenido al Juego de Adivinanza de Números!")
print("Estoy pensando en un número entre 1 y 100.")

target_number = generar_numero()
guess_counter = 0

while True:
    try:
        user_input = input("Introduce tu suposición: ")
        guess = int(user_input)
    except ValueError:
        print("Entrada inválida. Por favor, introduce un número.")
        continue

    guess_counter += 1

    if guess < target_number:
        print("¡Demasiado bajo! Inténtalo de nuevo.")
    elif guess > target_number:
        print("¡Demasiado alto! Inténtalo de nuevo.")
    else:
        print(f"¡Felicidades! Adivinaste el número {target_number} en {guess_counter} intentos!")
        break
```

## Generar Código de Calculadora Completa

### Subtask:
Generar el código Python completo para una calculadora de operaciones básicas, incluyendo funciones para suma, resta, multiplicación y división; un menú de opciones interactivo dentro de un bucle `while`; el uso de `if-elif-else` para seleccionar la operación; implementación de `try-except` para manejar errores como la división por cero y la entrada inválida; el uso de `continue` para regresar al menú principal después de cada operación; y una opción para salir del programa usando `break`.


```python
def suma(a, b):
    """Realiza la operación de suma."""
    return a + b

def resta(a, b):
    """Realiza la operación de resta."""
    return a - b

def multiplicacion(a, b):
    """Realiza la operación de multiplicación."""
    return a * b

def division(a, b):
    """Realiza la operación de división, manejando la división por cero."""
    if b == 0:
        raise ZeroDivisionError("Error: No se puede dividir por cero.")
    return a / b

print("¡Bienvenido a la Calculadora Simple!")

while True:
    print("\n--- Menú de Opciones ---")
    print("1. Suma")
    print("2. Resta")
    print(
"3. Multiplicación")
    print("4. División")
    print("5. Salir")

    choice = input("Por favor, elige una opción (1-5): ")

    if choice == '5':
        print("¡Gracias por usar la calculadora! ¡Hasta luego!")
        break
    elif choice in ('1', '2', '3', '4'):
        try:
            num1_str = input("Introduce el primer número: ")
            num2_str = input("Introduce el segundo número: ")

            num1 = float(num1_str)
            num2 = float(num2_str)

            result = None
            operation_name = ""

            if choice == '1':
                result = suma(num1, num2)
                operation_name = "Suma"
            elif choice == '2':
                result = resta(num1, num2)
                operation_name = "Resta"
            elif choice == '3':
                result = multiplicacion(num1, num2)
                operation_name = "Multiplicación"
            elif choice == '4':
                result = division(num1, num2)
                operation_name = "División"

            print(f"El resultado de la {operation_name} es: {result}")

        except ValueError:
            print("Entrada inválida. Por favor, introduce solo números válidos.")
        except ZeroDivisionError as e:
            print(e)
        except Exception as e:
            print(f"Ocurrió un error inesperado: {e}")
        finally:
            # Continue to the main menu after an operation or error
            continue
    else:
        print("Opción inválida. Por favor, elige una opción del 1 al 5.")
        continue
```

```python
def mostrar_menu():
    print("\n--- SISTEMA DE GESTIÓN DE TAREAS ---")
    print("1. Agregar tarea")
    print("2. Eliminar tarea")
    print("3. Mostrar todas las tareas")
    print("4. Salir")

def gestionar_tareas():
    # 1. Crear una lista vacía
    tareas = []

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        # 4. Usar if-elif-else para manejar el menú
        if opcion == "1":
            # 2. Función/Lógica para agregar
            nueva_tarea = input("Escriba la descripción de la tarea: ")
            tareas.append(nueva_tarea)
            print(f"Tarea '{nueva_tarea}' agregada con éxito.")

        elif opcion == "2":
            # 2 y 6. Función para eliminar con manejo de errores (try-except)
            try:
                if not tareas:
                    print("La lista está vacía, no hay nada que eliminar.")
                    continue

                for i, tarea in enumerate(tareas):
                    print(f"{i}. {tarea}")

                indice = int(input("Ingrese el número de la tarea a eliminar: "))
                eliminada = tareas.pop(indice)
                print(f"Tarea '{eliminada}' eliminada.")

            except (ValueError, IndexError):
                print("Error: Por favor, ingrese un número de índice válido.")

        elif opcion == "3":
            # 2 y 5. Función para mostrar usando bucle for
            print("\n--- LISTA ACTUAL DE TAREAS ---")
            if not tareas:
                print("No hay tareas pendientes.")
            else:
                for idx, tarea in enumerate(tareas, start=1):
                    print(f"{idx}. {tarea}")

        elif opcion == "4":
            # 7. Usar break para salir
            print("Saliendo del programa... ¡Hasta luego!")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecutar el programa
if __name__ == "__main__":
    gestionar_tareas()
```
